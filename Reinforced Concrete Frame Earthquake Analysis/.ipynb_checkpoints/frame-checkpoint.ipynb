```python
from openseespy.opensees import *
import opsvis as opsv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
wipe()
model('basic','-ndm',3, '-ndf', 6)
```

```python
# Dimension of frame: Units in kips, ft
width = 360.0
height = 144.0
# Create nodes
node(1, 0.0, 0.0, 0.0)
node(2, width, 0.0, 0.0)
node(3, 0.0, height, 0.0)
node(4, width, height, 0.0)
# Fix at base
fix(1, 1, 1, 1, 1, 1, 1)
fix(2, 1, 1, 1, 1, 1, 1)
```

```python
# MATERIAL DEFINITION
# # core concrete (confined) => uniaxialMaterial('Concrete01', matTag, fpc, epsc0, fpcu, epsU)
uniaxialMaterial('Concrete01', 1, -6.0, -0.004, -5.0, -0.014)
# # cover concrete (unconfined) => uniaxialMaterial('Concrete01', matTag, fpc, epsc0, fpcu, epsU)
uniaxialMaterial('Concrete01', 2, -5.0, -0.002, 0.0, -0.006)
# # Reinforcing steel => uniaxialMaterial('Steel01', matTag, Fy, E0, b)
    # b => Strain hardening ratio
uniaxialMaterial('Steel01', 3, 60.0, 30000.0, 0.01) 
```

```python
# Define cross-section for columns

colWidth = 15
colDepth = 24
c = 1.5
As = 0.60  # area of no. 7 bars

# Centroid location
y1 = colDepth / 2.0
z1 = colWidth / 2.0
```

```python
# Define fiber section => section('Fiber', secTag, '-GJ', GJ)
## patch('rect', matTag, numSubdivY, numSubdivZ, *crdsI, *crdsJ)
fib_sec_1 = [['section','Fiber', 1, '-GJ', 1.0e6],
             ['patch', 'rect', 1, 10, 1, c-y1, c-z1, y1-c, z1-c], #core
             ['patch', 'rect', 2, 10, 1, -y1, z1 - c, y1, z1], #top
             ['patch', 'rect', 2, 10, 1, -y1, -z1, y1, c - z1], #bottom
             ['patch', 'rect', 2, 2, 1, -y1, c - z1, c - y1, z1 - c], #left
             ['patch', 'rect', 2, 2, 1, y1 - c, c - z1, y1, z1 - c], #right
             ['layer', 'straight', 3, 3, As, y1 - c, z1 - c, y1 - c, c - z1], # left
             ['layer', 'straight', 3, 2, As, 0.0, z1 - c, 0.0, c - z1], # middle
             ['layer', 'straight', 3, 3, As, c - y1, z1 - c, c - y1, c - z1] # right
             ]

opsv.fib_sec_list_to_cmds(fib_sec_1) # defines fiber through 'opsvis'
opsv.plot_fiber_section(fib_sec_1)
plt.axis('equal')
plt.axhline(0,color='r')  # Z-axis
plt.axvline(0, color='r')  # Y-axis
plt.xlim(-z1 * 1.2, z1 * 1.2)
plt.ylim(-y1 * 1.2, y1 * 1.2)
plt.show()
```

```python
# Define column elements
geomTransf('PDelta', 1, 0, 1, 0)   # says 'Align my local y-axis with vector <0,1,0> in global space  

# Lobatto integration => beamIntegration('Lobatto', tag, secTag, N = number of integration points)
beamIntegration('Lobatto', 1, 1, 5)

# Create columns => element('forceBeamColumn', eleTag, *eleNodes, transfTag, integrationTag)
element('forceBeamColumn', 1, *[1, 3], 1, 1)
element('forceBeamColumn', 2, *[2, 4], 1, 1)
```

```python
# Define beam element
geomTransf('Linear', 2)
# Create beam => element('elasticBeamColumn', eleTag, *eleNodes, Area, E_mod, Iz, transfTag)
element('elasticBeamColumn', 3, *[3, 4], 360.0, 4030.0, 8640.0, 2)
```

```python
opsv.plot_model()
```

```python
h = 15
b = 24
ele_shapes = {1: ['rect', [b]],
              2: ['rect', [b]],
              3: ['rect', [b]]}

opsv.plot_extruded_shapes_3d(ele_shapes)
plt.show()
```

```python
ops.wipe()

ops.model('basic', '-ndm', 3, '-ndf', 6)

b = 0.2
h = 0.4

A, Iz, Iy, J = 0.04, 0.0010667, 0.0002667, 0.01172

E = 25.0e6
G = 9615384.6

# Lx, Ly, Lz = 4., 3., 5.
Lx, Ly, Lz = 4., 4., 4.

ops.node(1, 0., 0., 0.)
ops.node(2, 0., 0., Lz)
ops.node(3, Lx, 0., Lz)
ops.node(4, Lx, Ly, Lz)

ops.fix(1, 1, 1, 1, 1, 1, 1)

lmass = 200.

ops.mass(2, lmass, lmass, lmass, 0.001, 0.001, 0.001)
ops.mass(3, lmass, lmass, lmass, 0.001, 0.001, 0.001)
ops.mass(4, lmass, lmass, lmass, 0.001, 0.001, 0.001)

gTTagz = 1
gTTagx = 2
gTTagy = 3

coordTransf = 'Linear'
ops.geomTransf(coordTransf, gTTagz, 0., -1., 0.)
ops.geomTransf(coordTransf, gTTagx, 0., -1., 0.)
ops.geomTransf(coordTransf, gTTagy, 1., 0., 0.)

ops.element('elasticBeamColumn', 1, 1, 2, A, E, G, J, Iy, Iz, gTTagz)
ops.element('elasticBeamColumn', 2, 2, 3, A, E, G, J, Iy, Iz, gTTagx)
ops.element('elasticBeamColumn', 3, 3, 4, A, E, G, J, Iy, Iz, gTTagy)

Ew = {}

Px = -4.e1
Py = -2.5e1
Pz = -3.e1

ops.timeSeries('Constant', 1)
ops.pattern('Plain', 1, 1)
ops.load(4, Px, Py, Pz, 0., 0., 0.)

ops.constraints('Transformation')
ops.numberer('RCM')
ops.system('BandGeneral')
ops.test('NormDispIncr', 1.0e-6, 6, 2)
ops.algorithm('Linear')
ops.integrator('LoadControl', 1)
ops.analysis('Static')
ops.analyze(1)


opsv.plot_model()
```

```python
ele_shapes = {1: ['circ', [h]],
              2: ['rect', [b, h]],
              3: ['I', [b, h, b/10., h/6.]]}
opsv.plot_extruded_shapes_3d(ele_shapes)

plt.show()
```
